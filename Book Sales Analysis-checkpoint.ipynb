```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(
    r"C:\Users\mdaki\OneDrive\Desktop\Data Analyst projects\Dataset\Books_Data_Clean\Books_Data_Clean.csv",
    sep=",",
    engine="python",
    encoding="latin-1",
    on_bad_lines="skip"
)
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df[df["Publishing Year"] > 1900]
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset = "Book Name",inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

### Publishing Year Distribution – Insight

- Majority of books were published between 2000 and 2020.
- There is a noticeable increase in publications after 1990.
- Recent years show higher publishing activity, indicating growth in the book market.
- Older books (before 2000) represent a smaller proportion of total sales.

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
```

### Genre Distribution – Insight

- Fiction and Non-Fantasy are the most represented genres.
- Fiction and Children categories have fewer titles.
- High representation in Fiction suggests strong publisher focus on entertainment content.

```python
df["genre"].value_counts().plot(kind = "bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of Books in Each Genre")
plt.show()
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending = False)
```

### Rating Count by Genre – Insight

- Fiction shows higher median rating counts compared to other genres.
- Some genres display significant outliers, indicating a few highly popular books.
- Non-fiction genres like children have lower engagement variability.

```python
sns.boxenplot(x = "genre", y = "Book_ratings_count", data = df)
plt.xlabel("Genre")
plt.ylabel("Book Rating Count")
plt.title("Box Plot of Book Rating Count for Each Genre")
plt.show()
```

### Sales Price vs Units Sold – Insight

- There appears to be a slight negative correlation between price and units sold.
- Lower-priced books generally achieve higher sales volume.
- Premium-priced books tend to sell fewer copies but may generate high revenue per unit.

```python
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Scatter Plot of Sales Price Vs Units Sold")
plt.show()
```

### Average Rating vs Rating Count – Insight

- Books with high rating counts tend to maintain ratings of 71%.
- Extremely high ratings with low rating counts may indicate limited audience reach.
- Popular books generally balance both strong ratings and high engagement.

```python
plt.scatter(df["Book_average_rating"], df["Book_ratings_count"])
plt.xlabel("Book_average_rating")
plt.ylabel("Book_ratings_count")
plt.title("Scatter Plot of Book Average Rating Vs Book Rating Count")
plt.show()
```

```python
total_gross_sales_by_author = df.groupby("Author")["gross sales"].sum()
```

### Gross Sales by Author – Insight

- A small number of authors contribute disproportionately to total revenue.
- Top-performing authors significantly outperform mid-tier authors.
- Revenue concentration suggests brand value of established authors.

```python
total_gross_sales_by_author.sort_values(ascending = False).head(20).plot(kind = "bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales for Each Author")
plt.show()
```

### Units Sold by Author Rating – Insight

- Higher-rated authors generally show higher median sales.
- Variability in sales increases for highly rated authors.
- Strong author reputation positively impacts commercial performance.

```python
sns.boxplot(x = "Author_Rating", y = "units sold", data = df)
plt.xlabel("Author Rating")
plt.ylabel("Units Sold")
plt.title("Box Plot of Units Sold for Each Author Rating")
plt.show()
```

### Sales Trend Over Time – Insight

- Sales show an upward trend over the years.
- Significant growth observed after 2015.
- Seasonal spikes may indicate holiday-driven purchases.

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line", marker="o")
plt.xlabel("Publishing Year")
plt.ylabel("Total Units Sold")
plt.title("Total Units Sold Over the Year")
plt.show()
```
