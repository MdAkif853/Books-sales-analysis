```python
numeric_cols = ["BudgetUSD", "US_BoxOfficeUSD", "Global_BoxOfficeUSD", "Opening_Day_SalesUSD", "One_Week_SalesUSD", "IMDbRating", "RottenTomatoesScore", "NumVotesIMDb", "NumVotesRT"]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
for col in numeric_cols:
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
categorical_col = ["Genre", "Country", "Director", "LeadActor"]
```

```python
for col in categorical_col:
    df[col].value_counts().head(20).plot(kind="bar")
    plt.title(f"Top 20 {col} counts")
    plt.show()
```

```python
df.columns
```

```python
sns.scatterplot(data = df, x = "BudgetUSD", y = "Global_BoxOfficeUSD", alpha = 0.3)
plt.title("Budge vs Global Box Office")
plt.xscale("log")
plt.yscale("log")
plt.show()
```

```python
sns.scatterplot(data = df, x = "IMDbRating", y = "RottenTomatoesScore", alpha = 0.3)
plt.title("IMDb vs Rotten Tomatoes Ratings")
plt.show()
```

```python
genre_year_stats = df.groupby(["Genre", "ReleaseYear"]).agg({
    "Global_BoxOfficeUSD" : "mean", 
    "BudgetUSD" : "mean",
    "IMDbRating" : "mean",
    "RottenTomatoesScore" : "mean"
}).reset_index()
```

```python
genre_year_stats
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=genre_year_stats.query("Genre == 'Action'"), x = "ReleaseYear", y = "Global_BoxOfficeUSD", hue = "Genre")
plt.title("Average Global Box Office by Genre Over Year")
plt.show()
```

```python
Top_Actors = df.groupby("LeadActor")["IMDbRating"].mean().sort_values(ascending=False).head(10)
```

```python
Top_Actors
```

```python
Top_Actors.plot(kind = "bar")
plt.title("Top 10 Actors by Avg IMDb Rating")
plt.show()
```
